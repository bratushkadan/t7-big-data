```python
import pandas as pd

data = pd.read_csv("datasets/pr3/insurance.csv")
```

```python
print(data.head())
```

```python
# 2. С помощью метода describe() посмотреть статистику по данным. Сделать выводы.
print(data.describe())
# Выводы: средний и медианный возрасты - 39 лет
# Медианный и средний индексы массы тела 30 – ожирение
# В среднем один ребенок
# средний размер страхового вознаграждения – $ 13,270; медианный – $ 9,382
```

```python
# 3. Построить гистограммы для числовых показателей. Сделать выводы.
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Распределение числовых показателей')

data['age'].plot(kind='hist', ax=axs[0, 0])
axs[0, 0].set_title('Возраст')
axs[0, 0].set_xlabel('Возраст')

data['bmi'].plot(kind='hist', ax=axs[0, 1])
axs[0, 1].set_title('Индекс массы тела')
axs[0, 1].set_xlabel('BMI')

data['children'].plot(kind='hist', ax=axs[1, 0])
axs[1, 0].set_title('Количество детей')
axs[1, 0].set_xlabel('Количество детей')

data['charges'].plot(kind='hist', ax=axs[1, 1])
axs[1, 1].set_title('Страховые выплаты (Charges)')
axs[1, 1].set_xlabel('Страховые выплаты')

plt.subplots_adjust(hspace=0.5)
plt.show()
```
