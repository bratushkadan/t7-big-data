```python
# 3. Построить гистограммы для числовых показателей. Сделать выводы.
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Распределение числовых показателей')

data['age'].plot(kind='hist', ax=axs[0, 0])
axs[0, 0].set_title('Возраст')
axs[0, 0].set_xlabel('Возраст')

data['bmi'].plot(kind='hist', ax=axs[0, 1])
axs[0, 1].set_title('Индекс массы тела')
axs[0, 1].set_xlabel('BMI')

data['children'].plot(kind='hist', ax=axs[1, 0])
axs[1, 0].set_title('Количество детей')
axs[1, 0].set_xlabel('Количество детей')

data['charges'].plot(kind='hist', ax=axs[1, 1])
axs[1, 1].set_title('Страховые выплаты (Charges)')
axs[1, 1].set_xlabel('Страховые выплаты')

plt.subplots_adjust(hspace=0.5)
plt.show()
```

```python
# 4. Найти меры центральной тенденции и меры разброса для индекса массы
# тела (bmi) и расходов (charges). Отобразить результаты в виде текста и на
# гистограммах (3 вертикальные линии). Добавить легенду на графики.
# Сделать выводы.
# мода, среднее, медиана - меры центральной тенденции
bmi_mean = data['bmi'].mean()
bmi_median = data['bmi'].median()
# размах, стандартное отклонение, межквартильный размах - меры разброса (изменчивости)
bmi_std = data['bmi'].std()


charges_mean = data['charges'].mean()
charges_median = data['charges'].median()
charges_std = data['charges'].std()

fig, axs = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Распределение индекса массы тела (BMI) и страховых выплат (Charges)')

# Гистограмма для BMI
data['bmi'].plot(kind='hist', ax=axs[0], bins=30, color='lightblue', edgecolor='black')
axs[0].axvline(bmi_mean, color='red', linestyle='dashed', linewidth=2, label=f'Среднее: {bmi_mean:.2f}')
axs[0].axvline(bmi_median, color='green', linestyle='dashed', linewidth=2, label=f'Медиана: {bmi_median:.2f}')
axs[0].axvline(bmi_mean + bmi_std, color='purple', linestyle='dashed', linewidth=2, label=f'Стандартное отклонение: {bmi_std:.2f}')
axs[0].set_title('Индекс массы тела (BMI)')
axs[0].set_xlabel('BMI')
axs[0].legend()

# Гистограмма для Charges
data['charges'].plot(kind='hist', ax=axs[1], bins=30, color='lightblue', edgecolor='black')
axs[1].axvline(charges_mean, color='red', linestyle='dashed', linewidth=2, label=f'Среднее: {charges_mean:.2f}')
axs[1].axvline(charges_median, color='green', linestyle='dashed', linewidth=2, label=f'Медиана: {charges_median:.2f}')
axs[1].axvline(charges_mean + charges_std, color='purple', linestyle='dashed', linewidth=2, label=f'Стандартное отклонение: {charges_std:.2f}')
axs[1].set_title('Страховые выплаты (Charges)')
axs[1].set_xlabel('Charges')
axs[1].legend()
plt.subplots_adjust(wspace=0.4)
plt.show()

# Вывод результатов в виде текста
print("\nИндекс массы тела:")
print(f"Среднее значение: {bmi_mean:.2f}")
print(f"Медиана: {bmi_median:.2f}")
print(f"Стандартное отклонение: {bmi_std:.2f}")

print("\nСтраховые выплаты:")
print(f"Среднее значение: {charges_mean:.2f}")
print(f"Медиана: {charges_median:.2f}")
print(f"Стандартное отклонение: {charges_std:.2f}")
```

```python
# 6. Используя признак imb, проверить, выполняется ли центральная предельная теорема.
# Использовать различные длины выборок n. Количество выборок = 300.
# Вывести результат в виде гистограмм. Найти стандартное отклонение и среднее
# для полученных распределений.

# генератор выборки
def sample_means(data, n, num_samples):
    samples = np.random.choice(data, (num_samples, n))
    return samples.mean(axis=1)

# признак индекса массы тела
bmi_data = data['bmi'].values

# длина выборов и количество выборок
n_values = [10, 30, 50, 250]
num_samples = 300

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Гистограммы средних значений BMI')

for i, n in enumerate(n_values):
    sample_means_data = sample_means(bmi_data, n, num_samples)
    
    row, col = divmod(i, 2)
    axs[row, col].hist(sample_means_data, bins=30, edgecolor='black', alpha=0.7)
    axs[row, col].set_title(f'n = {n}')
    axs[row, col].set_xlabel('Среднее значение BMI')
    axs[row, col].set_ylabel('Частота')
    
    mean = np.mean(sample_means_data)
    std = np.std(sample_means_data)
    
    # Добавим линию среднего значения и подписи для среднего и стандартного отклонения
    axs[row, col].axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Среднее: {mean:.2f}')
    axs[row, col].axvline(mean + std, color='purple', linestyle='dashed', linewidth=2, label=f'Стандартное отклонение: {std:.2f}')
    axs[row, col].axvline(mean - std, color='purple', linestyle='dashed', linewidth=2)
    axs[row, col].legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# 1. По мере увеличения размера выборки распределение средних значений приближается к нормальному (колонообразному)
# 2. «Чем больше длина выборок n и чем меньше дисперсия исследуемых данных, тем меньше будет стандартная ошибка среднего».
# Действительно, как можем видеть, при большем n (размере выборки) меньше и стандартное отклонение (меньше разброс).
```

```python
# 7. Построить 95% и 99% доверительный интервал для среднего значения
# расходов и среднего значения индекса массы тела.
# Выбираем данные по BMI и Charges
import scipy.stats as stats

bmi_data = data['bmi']
charges_data = data['charges']

# расчет доверительного интервала
def confidence_interval(data, confidence):
    mean = np.mean(data)
    # стандартная ошибка среднего значения
    std_err = stats.sem(data)
    n = len(data)
    z = stats.norm.ppf((1 + confidence) / 2)
    
    h = z * std_err
    return (mean - h, mean + h)


bmi_95_conf_interval = confidence_interval(bmi_data, 0.95)
bmi_99_conf_interval = confidence_interval(bmi_data, 0.99)
charges_95_conf_interval = confidence_interval(charges_data, 0.95)
charges_99_conf_interval = confidence_interval(charges_data, 0.99)

print(f'95% доверительный интервал для среднего значения BMI: ({bmi_95_conf_interval[0]:.2f}, {bmi_95_conf_interval[1]:.2f})')
print(f'99% доверительный интервал для среднего значения BMI: ({bmi_99_conf_interval[0]:.2f}, {bmi_99_conf_interval[1]:.2f})')
print(f'95% доверительный интервал для среднего значения Charges: ({charges_95_conf_interval[0]:.2f}, {charges_95_conf_interval[1]:.2f})')
print(f'99% доверительный интервал для среднего значения Charges: ({charges_99_conf_interval[0]:.2f}, {charges_99_conf_interval[1]:.2f})')

# Вывод: с вероятностью 95/99% истинное среднее значение BMI/Charges в популяции лежит в выведенном интервале.
```

```python
# 8. Проверить распределения следующих признаков на нормальность:
# индекс массы тела, расходы. Сформулировать нулевую и альтернативную
# гипотезы. Для каждого признака использовать KS-тест и q-q plot. Сделать
# выводы на основе полученных p-значений.

from scipy.stats import kstest

bmi_data = data['bmi']
charges_data = data['charges']

# run KS-test and build Q-Q plot
def check_normality(data, label):
    ks_statistic, p_value = kstest(data, 'norm', args=(np.mean(data), np.std(data)))
    
    plt.figure(figsize=(6, 6))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f'Q-Q plot for {label}')
    plt.show()
    
    return ks_statistic, p_value

# проверка нормальности для BMI
bmi_ks_statistic, bmi_p_value = check_normality(bmi_data, 'BMI')

# проверка нормальности для Charges
charges_ks_statistic, charges_p_value = check_normality(charges_data, 'Charges')

print(f'KS-тест для BMI: Статистика = {bmi_ks_statistic}, p-значение = {bmi_p_value}')
print(f'KS-тест для Charges: Статистика = {charges_ks_statistic}, p-значение = {charges_p_value}')
# Вывод:
# - p-значение для BMI ≈ 0.32 > 0.05 – нулевая гипотеза не отвергается - распределение BMI близко к нормальному
# - Q-Q plot график для BMI похож скорее на идеальную прямую – данные следуют нормальному закону распределения
# - p-значение для Charges ≈ 4.4e-42 < 0.05 - нулевая гипотеза отвергается - распределение не является нормальным
# - Q-Q plot график для Charget также не похож на прямую - данные не следуют нормальному закону распределения
```
